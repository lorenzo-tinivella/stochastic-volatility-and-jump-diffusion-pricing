```python
import sys
import os
sys.path.append(os.path.abspath('../src'))

import numpy as np
import matplotlib.pyplot as plt
# 
from models.merton_jump import MertonJumpDiffusion
```

```python
S0 = 100
r = 0.05
sigma = 0.2

lam = 0.75
mu_j = -0.1
sigma_j = 0.3

T = 1
K = 100
```

```python
model = MertonJumpDiffusion(S0, r, sigma, lam, mu_j, sigma_j)
```

```python
price = model.call_price_mc(K, T)

print("Call price (Merton Jump Diffusion):", price)
```

```python
ST = model.simulate_terminal_price(T, 100000)

plt.hist(np.log(ST), bins=100)

plt.title("Log-price distribution under Merton model")
plt.xlabel("log(ST)")
plt.ylabel("Frequency")

plt.show()
```

```python
from models.black_scholes import BlackScholes

bs = BlackScholes(S0, r, sigma)

ST_bs = bs.simulate(T, 100000)
ST_merton = model.simulate_terminal_price(T, 100000)

plt.hist(ST_bs, bins=100, alpha=0.5, label="Black-Scholes")
plt.hist(ST_merton, bins=100, alpha=0.5, label="Merton")

plt.xlim(0,300)

plt.legend()
plt.title("Terminal price distribution comparison")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```
